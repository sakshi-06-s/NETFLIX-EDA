```python
df_clean["cast"]=df_clean["cast"].fillna("unknown")
```

```python
##INSIGHTS SUMMARY
```

1. Content Type Distribution

Netflix has more Movies than TV Shows, showing that the platform is more movie-heavy. However, TV Shows play an important role in user retention and long-term engagement.

2. Content Growth Over Time

Netflix content increased sharply after 2015, with peak growth between 2017–2019. This shows aggressive global expansion and rapid content acquisition strategy.

3. Top Countries Producing Content

The United States dominates Netflix content production. Countries like India, UK, Canada, and France are emerging contributors, showing Netflix’s shift toward global and regional content expansion.

 4. Genre Distribution

Drama and International Movies dominate Netflix’s library, followed by Comedy and Action. This indicates Netflix focuses on high-engagement, globally appealing storytelling formats.

5. Ratings Distribution

Most content is rated TV-MA and TV-14, meaning Netflix primarily targets adult and teen audiences. Kids content is relatively small in proportion.

6. Movie Duration Distribution

Most movies fall between 90–120 minutes, which is standard industry length. Netflix follows conventional film duration patterns optimized for viewer engagement.

7. TV Show Duration Distribution

Most TV shows have 1–2 seasons, showing Netflix preference for short or limited series formats instead of long-running shows.

8. Missing Data Pattern

Missing values exist mainly in duration,ratings, and date_added. This is common in large-scale scraped datasets and has been handled using imputation and labeling techniques.

9. Ratings vs Content Type

Both Movies and TV Shows are dominated by TV-MA and TV-14 ratings, showing consistent focus on mature audiences across all content types.

10. Duration vs Rating Analysis

Duration is not strictly dependent on rating, but mature-rated content shows wider variation in length, while kids content is shorter and more consistent.

11. Movies vs TV Shows Duration Pattern

Movies show a tight distribution around 90–120 minutes, while TV Shows are heavily skewed toward 1 season with a long tail of longer series.

12. Monthly + Yearly Content Pattern

Netflix content is added in batch releases rather than evenly throughout the year, with clear spikes in certain months and strong growth over time.

```python
##FINAL OVERALL INSIGHT
```

Netflix is a global, adult-focused streaming platform that:

-prioritizes movies over TV shows
-focuses on drama and international content
-expands aggressively into global markets
-prefers short-form TV series and standard-length movies
-releases content in structured batches rather than continuous flow
